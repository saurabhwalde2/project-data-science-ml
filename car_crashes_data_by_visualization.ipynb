Since, ‘speeding’, ‘alcohol’, ‘not_distracted’,’no_previous’ are part a breakdown of the ‘total’ column, it would be beneficial to compare percentages or porportions across each state as shown in the bad drivers dataset. To add the proportions, divide each column by the ‘total’ column.

percent_speeding – Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding.

percent_alcohol – Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired.

percent_not_distracted – Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted.

percent_no_previous – Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents**

```python
car_crashes.shape
```

```python
car_crashes.head(5)
```

```python
car_crashes['percent_speeding'] = (car_crashes['speeding']/car_crashes['total'])*100
car_crashes['percent_alcohol'] = (car_crashes['alcohol']/car_crashes['total'])*100
car_crashes['percent_not_distracted'] = (car_crashes['not_distracted']/car_crashes['total'])*100
car_crashes['percent_no_previous'] = (car_crashes['no_previous']/car_crashes['total'])*100
```

```python
car_crashes.head(10)
```

```python
sns.pairplot(car_crashes)
```

```python
data = car_crashes.sort_values(by='total', ascending=False)
print(data[['abbrev','total','alcohol']].head())
```

```python
plt.figure(figsize=(10,15))
sns.barplot(data=data,x='total',y='abbrev')
sns.barplot(data=data,x='alcohol',y='abbrev', color='grey')

# Set the title of the plot
plt.title('Barplot of Total Collisionsn(Grey bar indicates number of people intoxicated)')
# Set the x-axis label
plt.xlabel('Number of Drivers in Fatal Collisions')
# Set the y-axis label
plt.ylabel('State') ;
```

**South Carolina, North Dakota and West Virginia have approximately 24 people involed in fatal collisions. Approximately 10 individauls out of 24 in South Carolina and North Dakota had caused alcohol related fatal accidents.**

Lets try to reproduce some graphs from here: https://fivethirtyeight.com/features/which-state-has-the-worst-drivers/ The author is trying to answer the question about “Which state has the most bad drivers?” Author also goes on to clarify that just because a state has high accidents does not mean it has the most drivers. It could be contibuted to the number of drivers, amount of miles driven etc, population etc.

```python
data_alcohol = car_crashes.sort_values(by='alcohol', ascending=False)
data_alcohol = data_alcohol.head(10)

data_percent_alcohol = car_crashes.sort_values(by='percent_alcohol', ascending=False)
data_percent_alcohol = data_percent_alcohol.head(10)


data_speeding = car_crashes.sort_values(by='speeding', ascending=False)
data_speeding = data_speeding.head(10)

data_percent_speeding = car_crashes.sort_values(by='percent_speeding', ascending=False)
data_percent_speeding = data_percent_speeding.head(10)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=data_alcohol,x='alcohol',y='abbrev')
plt.title('Top 10 Alcohol Related Collisions')
sns.barplot(data=data_percent_alcohol,x='percent_alcohol',y='abbrev')
plt.title('Top 10 Portion of Alcohol Related Collisions');

#note we are not getting both the plot thats way we are using subplots
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first subplot (Top 10 Alcohol Related Collisions)
sns.barplot(data=data_alcohol, x='alcohol', y='abbrev', ax=ax[0])
ax[0].set_title('Top 10 Alcohol Related Collisions')

# Plot the second subplot (Top 10 Portion of Alcohol Related Collisions)
sns.barplot(data=data_percent_alcohol, x='percent_alcohol', y='abbrev', ax=ax[1])
ax[1].set_title('Top 10 Portion of Alcohol Related Collisions')

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()  # Display the plots
```

**Even though North Dakota has the highest number of Alcohol related fatalities, it is Montana that has the highest porportion of Alcohol related fatalities.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first subplot (Top 10 States with Speeding Related Collisions)
sns.barplot(data=data_speeding, x='speeding', y='abbrev', ax=ax[0])
ax[0].set_title('Top 10 States with Speeding Related Collisions')

# Plot the second subplot (Top 10 States with Portion of Speeding Related Collisions)
sns.barplot(data=data_percent_speeding, x='percent_speeding', y='abbrev', ax=ax[1])
ax[1].set_title('Top 10 States with Portion of Speeding Related Collisions')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()  # Display the plots
```

**It is interesting to note that Hawaii has the highest number of speeding related collisions. In addition, Hawaii also has the highest proportion of speeding related fatal collision.**
